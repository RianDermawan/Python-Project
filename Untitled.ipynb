```python
import pandas as pd
from datetime import datetime
```

```python
# read excel file
dataku = pd.read_excel('dataku.xlsx')
```

```python
# view data
dataku
```

```python
# Because in column 'Date' the type data is string, convert type data to sring
dataku['Date'] = pd.to_datetime(dataku['Date'], format = '%d-%m-%Y')
```

```python
# add column year and month
dataku['Year'] = pd.DatetimeIndex(dataku['Date']).year
dataku['Month'] = pd.DatetimeIndex(dataku['Date']).month
```

```python
# view new data
dataku
```

```python
# if column 'RR' <= 80 convert to NaN
dataku.loc[dataku['RR']<=80, 'RR'] = float('NaN')
```

```python
# view new data
dataku
```

```python
# group data by column 'Month' and 'Year', next count the data 
data_baru = dataku.groupby(['Month', 'Year'], as_index=False).count()
```

```python
data_baru
```

```python
# slicing data
data_baru = data_baru[['Month', 'Year', 'RR']]
```

```python
# export to excel file
data_baru.to_excel('more than 80.xlsx', index=False)
```

```python
# group data by column 'Month' and 'Year', next calculate max the data 
dataku.groupby(['Month', 'Year'], as_index=False).max()
```

```python
# define new variabel
data_baru2 = dataku.groupby(['Month', 'Year'], as_index=False).max()
```

```python
# slicing data
data_baru2 = data_baru2[['Month', 'Year', 'ff_x', 'ddd_x']]
```

```python
# view new data
data_baru2
```

```python
# export to excel file
data_baru2.to_excel('max.xlsx', index=False)
```
